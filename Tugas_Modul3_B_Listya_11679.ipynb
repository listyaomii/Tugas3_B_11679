```python
import pandas as pd
import numpy as np

df_ipk = pd.read_csv(r'dataset_regresi_IPK.csv')
df_ipk.head(10)
```

```python
from pandas.api.types import is_numeric_dtype
def  remove_outlier(df_in):
    for col_name in list(df_in.columns):
        if is_numeric_dtype(df_in[col_name]):
            q1 = df_in[col_name].quantile(0.25)
            q3 = df_in[col_name].quantile(0.75)

            iqr = q3-q1
            batas_atas = q3 + (1.5 * iqr)
            batas_bawah = q1 - (1.5 * iqr)

            df_out = df_in.loc[(df_in[col_name] >= batas_bawah) & (df_in[col_name] <= batas_atas)]
    return df_out

df_ipk_clean = remove_outlier(df_ipk2)
print("Jumlah baris DataFrame sebelum dibuang outliner" , df_ipk2.shape[0])
print("Jumlah baris DataFrame sesudah dibuang outliner", df_ipk_clean.shape[0])

df_ipk_clean.IPK.plot(kind='box', vert=True)

plt.gca().invert_yaxis()
plt.show()
```

```python
print("data null \n", df_ipk_clean.isnull().sum())
print("data kosong \n", df_ipk_clean.empty)
print("data nan \n", df_ipk_clean.isna().sum())
```

```python
from sklearn.model_selection import train_test_split

X_regress = df_ipk_clean.drop('IPK', axis=1)
y_regress = df_ipk_clean.IPK

X_train_ipk, X_test_ipk, y_train_ipk, y_test_ipk = train_test_split(X_regress, y_regress,
                                                                    test_size=0.25, random_state=79)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_absolute_error, mean_squared_error

pipe_Ridge = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', Ridge())
])

param_grid_Ridge = {
    'reg__alpha': [0.01,0.1,1,10,100],
    'feature_selection__k': np.arange(1,20)
}

GSCV_RR = GridSearchCV(pipe_Ridge, param_grid_Ridge, cv=5, scoring='neg_mean_squared_error', error_score='raise')

GSCV_RR.fit(X_train_ipk, y_train_ipk)

print("Best model:{}".format(GSCV_RR.best_estimator_))
print("Ridge best parameter:{}".format(GSCV_RR.best_params_))

print("Koefisien/bobot:{}".format(GSCV_RR.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bobot:{}".format(GSCV_RR.best_estimator_.named_steps['reg'].intercept_))

Ridge_predict = GSCV_RR.predict(X_test_ipk)

mse_Ridge = mean_squared_error(y_test_ipk, Ridge_predict)
mae_Ridge = mean_absolute_error(y_test_ipk, Ridge_predict)

print("Ridge Mean Squared Error (MSE): {}".format(mse_Ridge))
print("Ridge Mean Absolute Aerror (MAE):{}".format(mae_Ridge))
print("Ridge Root Mean Squared Error:{}".format(np.sqrt(mse_Ridge)))
```

```python
df_result = pd.DataFrame(y_test_ipk, columns=['IPK'])
df_result = pd.DataFrame(y_test_ipk)
df_result['Ridge Prediction'] = Ridge_predict

df_result['Selisih_IPK_RR'] = df_result['Ridge Prediction'] - df_result['IPK']

df_result.head()
```

```python
df_result.describe()
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_absolute_error, mean_squared_error

pipe_Lasso = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', Lasso(max_iter=1000))
])

param_grid_Lasso = {
    'reg__alpha': [0.01,0.1,1,10,100],
    'feature_selection__k': np.arange(1,20)
}

GSCV_Lasso = GridSearchCV(pipe_Lasso, param_grid_Lasso, cv=5, scoring='neg_mean_squared_error', error_score='raise')

GSCV_Lasso.fit(X_train_ipk, y_train_ipk)

print("Best model:{}".format(GSCV_Lasso.best_estimator_))
print("Lasso best parameter:{}".format(GSCV_Lasso.best_params_))

print("Koefisien/bobot:{}".format(GSCV_Lasso.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bobot:{}".format(GSCV_Lasso.best_estimator_.named_steps['reg'].intercept_))


Lasso_predict = GSCV_Lasso.predict(X_test_ipk)

mse_Lasso = mean_squared_error(y_test_ipk, Lasso_predict)
mae_Lasso = mean_absolute_error(y_test_ipk, Lasso_predict)

print("Lasso Mean Squared Error (MSE): {}".format(mse_Lasso))
print("Lasso Mean Absolute Aerror (MAE):{}".format(mae_Lasso))
print("Lasso Root Mean Squared Error:{}".format(np.sqrt(mse_Lasso)))
```

```python
df_result['Lasso Prediction'] = Lasso_predict
df_result = pd.DataFrame(y_test_ipk)
df_result['Lasso Prediction'] = Lasso_predict

df_result['Selisih_IPK_LR'] = df_result['Lasso Prediction'] - df_result['IPK']
df_result.head()
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_absolute_error, mean_squared_error

pipe_SVR = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', SVR(kernel='linear'))
])

param_grid_SVR = {
    'reg__C': [0.01, 0.1, 1, 10, 100],
    'reg__epsilon': [0.1, 0.2, 0.5, 1],
    'feature_selection__k': np.arange(1,20)
}

GSCV_SVR = GridSearchCV(pipe_SVR, param_grid_SVR, cv=5, scoring='neg_mean_squared_error', error_score='raise')
GSCV_SVR.fit(X_train_ipk, y_train_ipk)

print("Best model:{}".format(GSCV_SVR.best_estimator_))
print("SVR best parameter:{}".format(GSCV_SVR.best_params_))

print("Koefisien/bobot:{}".format(GSCV_SVR.best_estimator_.named_steps['reg'].coef_))
print("Intercept/bobot:{}".format(GSCV_SVR.best_estimator_.named_steps['reg'].intercept_))


SVR_predict = GSCV_SVR.predict(X_test_ipk)

mse_SVR = mean_squared_error(y_test_ipk, SVR_predict)
mae_SVR = mean_absolute_error(y_test_ipk, SVR_predict)

print("SVR Mean Squared Error (MSE): {}".format(mse_SVR))
print("SVR Mean Absolute Aerror (MAE):{}".format(mae_SVR))
print("SVR Root Mean Squared Error:{}".format(np.sqrt(mse_SVR)))
```

```python
df_result['SVR Prediction'] = SVR_predict
df_result = pd.DataFrame(y_test_ipk)
df_result['SVR Prediction'] = SVR_predict

df_result['Selisih_IPK_SVR'] = df_result['SVR Prediction'] - df_result['IPK']
df_result.head()
```

```python
df_result.describe()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_absolute_error, mean_squared_error

pipe_GBR = Pipeline(steps=[
            ('scale', StandardScaler()),
            ('feature_selection', SelectKBest(score_func=f_regression)),
            ('reg', GradientBoostingRegressor())
])


param_grid_GBR = {
    'reg__n_estimators': [100, 200, 300],
    'reg__learning_rate': [0.01, 0.1, 0.2],
    'reg__max_depth': [3, 4, 5],
    'feature_selection__k': np.arange(1,20)
}

GSCV_GBR = GridSearchCV(pipe_GBR, param_grid_GBR, cv=5, scoring='neg_mean_squared_error')
GSCV_GBR.fit(X_train_ipk, y_train_ipk)

print("Best model:{}".format(GSCV_GBR.best_estimator_))
print("SVR best parameter:{}".format(GSCV_GBR.best_params_))

print("Feature Importance: {}".format(GSCV_GBR.best_estimator_.named_steps['reg'].feature_importances_))

GBR_predict = GSCV_GBR.predict(X_test_ipk)

mse_GBR = mean_squared_error(y_test_ipk, GBR_predict)
mae_GBR = mean_absolute_error(y_test_ipk, GBR_predict)

print("GBR Mean Squared Error (MSE): {}".format(mse_GBR))
print("GBR Mean Absolute Aerror (MAE):{}".format(mae_GBR))
print("GBR Root Mean Squared Error:{}".format(np.sqrt(mse_GBR)))
```

```python
df_result['GBR Prediction'] = GBR_predict
df_result = pd.DataFrame(y_test_ipk)
df_result['GBR Prediction'] = GBR_predict

df_result['Selisih_IPK_GBR'] = df_result['GBR Prediction'] - df_result['IPK']
df_result.head()
```
